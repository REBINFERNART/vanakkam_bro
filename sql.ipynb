```python

import gradio as gr
import mysql.connector

# # Establish a connection to the MySQL database
# mydb = mysql.connector.connect(
#   host="localhost",
#   user="nethaji",
#   password="Sample@123",
#   database="rebin"
# )

# # Create a cursor
# mycursor = mydb.cursor()

# Define a function to insert data into the table
def insert_data(id, name, address):
    try:
        sql = "INSERT INTO rebin.guvi (id, name, address) VALUES (%s, %s, %s)"
        values = (id, name, address)
        mycursor.execute(sql, values)
        mydb.commit()
        return "Data inserted successfully!"
    except Exception as e:
        return f"Error: {str(e)}"

# Gradio interface
with gr.Interface(
    fn=insert_data,
    inputs=[gr.Number(label="ID"), gr.Textbox(label="Name"), gr.Textbox(label="Address")],
    outputs="text",
    title="Insert Data into rebin.guvi Table"
) as demo:
    demo.launch()
```
